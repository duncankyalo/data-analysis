```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.patches as patches

# Load the images
img1 = mpimg.imread('E:/GIS/MAPS/chirps may analysis.jpg')  # Replace with the path to your first image
img2 = mpimg.imread('E:/GIS/MAPS/wrf may analysis.jpg')  # Replace with the path to your second image

# Create a figure and axes for the combined image
fig, axs = plt.subplots(1, 2, figsize=(10, 10))  # Adjust figure size as needed

# Display the images in their respective axes
axs[0].imshow(img1)
axs[1].imshow(img2)

# Set titles for each axis (optional)
axs[0].set_title('Observed')
axs[1].set_title('Predicted')

# Remove axis ticks and labels (optional)
axs[0].axis('off')
axs[1].axis('off')

# Add brown frame around each image
for ax in axs:
    rect = patches.Rectangle((0, 0), 1, 1, linewidth=2, edgecolor='brown', facecolor='none', transform=ax.transAxes)
    ax.add_patch(rect)


# Adjust spacing between subplots (optional)
plt.tight_layout()

# Display the combined figure
plt.show()
```
